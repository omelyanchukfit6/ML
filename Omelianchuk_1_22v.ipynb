```python
import numpy as np
# –û—Ç—Ä–∏–º–∞–Ω–Ω—è –≤–µ—Ä—Å—ñ—ó –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ NumPy
numpy_version = np.__version__
print(f"NumPy version: {numpy_version}")
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 1
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –æ–¥–Ω–æ–≤–∏–º—ñ—Ä–Ω–∏–π –º–∞—Å–∏–≤ (–≤–µ–∫—Ç–æ—Ä) –∑—ñ –∑–Ω–∞—á–µ–Ω–Ω—è–º–∏ ùëí^ùë• –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ 
–Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞ x, –ø–æ—á–∏–Ω–∞—é—á–∏ –∑ 1, —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å –π–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è.
'''
# –í–∏–∑–Ω–∞—á–∏–º–æ –¥—ñ–∞–ø–∞–∑–æ–Ω –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–∏—Ö —á–∏—Å–µ–ª (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –≤—ñ–¥ 1 –¥–æ 10)
x_values = np.arange(1, 11)

# –û–±—á–∏—Å–ª–∏–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è e^x –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ x
exp_values = np.exp(x_values)

# –í–∏–≤–µ–¥–µ–º–æ —Ä–µ–∑—É–ª—å—Ç—É—é—á–∏–π –º–∞—Å–∏–≤
print("–ú–∞—Å–∏–≤ –∑–Ω–∞—á–µ–Ω—å e^x –¥–ª—è x –≤—ñ–¥ 1 –¥–æ 10:", exp_values)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 2
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤–æ–≤–∏–º—ñ—Ä–Ω–∏–π –º–∞—Å–∏–≤ (–º–∞—Ç—Ä–∏—Ü—é) —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å –π–æ–≥–æ —Ç–∞–Ω–≥–µ–Ω—Å–∞–º–∏ 
–∫–æ–∂–Ω–æ–≥–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞, –ø–æ—á–∏–Ω–∞—é—á–∏ –∑ 1, —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å –π–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è.
'''
# –í–∏–∑–Ω–∞—á–∏–º–æ –¥—ñ–∞–ø–∞–∑–æ–Ω –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–∏—Ö —á–∏—Å–µ–ª (–≤—ñ–¥ 1 –¥–æ 9)
x_values = np.arange(1, 10)

# –û–±—á–∏—Å–ª–∏–º–æ —Ç–∞–Ω–≥–µ–Ω—Å–∏ –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è
tan_values = np.tan(x_values)

# –°—Ç–≤–æ—Ä–∏–º–æ –¥–≤–æ–≤–∏–º—ñ—Ä–Ω–∏–π –º–∞—Å–∏–≤ 3x3 —ñ –∑–∞–ø–æ–≤–Ω–∏–º–æ –π–æ–≥–æ —Ç–∞–Ω–≥–µ–Ω—Å–∞–º–∏
matrix = tan_values.reshape(3, 3)

# –í–∏–≤–µ–¥–µ–º–æ —Ä–µ–∑—É–ª—å—Ç—É—é—á—É –º–∞—Ç—Ä–∏—Ü—é
print("–ú–∞—Ç—Ä–∏—Ü—è 3x3, –∑–∞–ø–æ–≤–Ω–µ–Ω–∞ —Ç–∞–Ω–≥–µ–Ω—Å–∞–º–∏ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–∏—Ö —á–∏—Å–µ–ª –≤—ñ–¥ 1 –¥–æ 9:")
print(matrix)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 3
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å –π–æ–≥–æ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -1 –¥–æ
1 —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -1 –¥–æ 1
matrix = np.random.uniform(-1, 1, (3, 3))

# –í–∏–≤–æ–¥–∏–º–æ –ø–æ—á–∞—Ç–∫–æ–≤—É –º–∞—Ç—Ä–∏—Ü—é
print("–ú–∞—Ç—Ä–∏—Ü—è 3x3 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -1 –¥–æ 1:")
print(matrix)

# –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —è–∫—ñ –µ–ª–µ–º–µ–Ω—Ç–∏ –º–∞—Ç—Ä–∏—Ü—ñ —î –Ω–µ–ø–∞—Ä–Ω–∏–º–∏
# –°–ø–æ—á–∞—Ç–∫—É –æ–∫—Ä—É–≥–ª–∏–º–æ –¥–æ —Ü—ñ–ª–∏—Ö —Ç–∞ –∑–∞–ª–∏—à–∏–º–æ –ª–∏—à–µ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞
odd_numbers = matrix[np.round(matrix).astype(int) % 2 != 0]

# –í–∏–≤–æ–¥–∏–º–æ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞
print("–ù–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –≤ –º–∞—Ç—Ä–∏—Ü—ñ:")
print(odd_numbers)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 4
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä–æ–º 4x4, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å –π–æ–≥–æ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -0.8
–¥–æ 0.8 —Ç–∞ –≤–∏–≤–µ–¥—ñ—Ç—å –π–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä–æ–º 4x4 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -0.8 –¥–æ 0.8
matrix = np.random.uniform(-0.8, 0.8, (4, 4))

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç—É—é—á–∏–π –º–∞—Å–∏–≤
print("–ú–∞—Ç—Ä–∏—Ü—è 4x4 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -0.8 –¥–æ 0.8:")
print(matrix)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 5
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤–∞ –æ–¥–Ω–æ–≤–∏–º—ñ—Ä–Ω–∏—Ö –º–∞—Å–∏–≤–∏ —Ä–æ–∑–º—ñ—Ä–æ–º 6, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—Ö –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏
—á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 5 –¥–æ 15 —Ç–∞ –≤–∏–∫–æ–Ω–∞–π—Ç–µ –Ω–∞ –Ω–∏—Ö –ø–æ–µ–ª–µ–º–µ–Ω—Ç–Ω—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó –¥–æ–¥–∞–≤–∞–Ω–Ω—è,
–≤—ñ–¥–Ω—ñ–º–∞–Ω–Ω—è —Ç–∞ –º–Ω–æ–∂–µ–Ω–Ω—è.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –¥–≤–∞ –æ–¥–Ω–æ–≤–∏–º—ñ—Ä–Ω–∏—Ö –º–∞—Å–∏–≤–∏ —Ä–æ–∑–º—ñ—Ä–æ–º 6 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 5 –¥–æ 15
array1 = np.random.randint(5, 16, 6)
array2 = np.random.randint(5, 16, 6)

# –í–∏–∫–æ–Ω—É—î–º–æ –ø–æ–µ–ª–µ–º–µ–Ω—Ç–Ω—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó
addition = array1 + array2
subtraction = array1 - array2
multiplication = array1 * array2

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–µ—Ä—à–∏–π –º–∞—Å–∏–≤:", array1)
print("–î—Ä—É–≥–∏–π –º–∞—Å–∏–≤:", array2)
print("–ü–æ–µ–ª–µ–º–µ–Ω—Ç–Ω–µ –¥–æ–¥–∞–≤–∞–Ω–Ω—è:", addition)
print("–ü–æ–µ–ª–µ–º–µ–Ω—Ç–Ω–µ –≤—ñ–¥–Ω—ñ–º–∞–Ω–Ω—è:", subtraction)
print("–ü–æ–µ–ª–µ–º–µ–Ω—Ç–Ω–µ –º–Ω–æ–∂–µ–Ω–Ω—è:", multiplication)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 6
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ 4x4, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—Ö –¥–æ–≤—ñ–ª—å–Ω–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -6 –¥–æ 6 —Ç–∞
–∑–Ω–∞–π–¥—ñ—Ç—å —ó—Ö –¥–æ–±—É—Ç–æ–∫.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ 4x4 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -6 –¥–æ 6
matrix1 = np.random.randint(-6, 7, (4, 4))
matrix2 = np.random.randint(-6, 7, (4, 4))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ –¥–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—å
product = np.dot(matrix1, matrix2)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–µ—Ä—à–∞ –º–∞—Ç—Ä–∏—Ü—è:")
print(matrix1)
print("\n–î—Ä—É–≥–∞ –º–∞—Ç—Ä–∏—Ü—è:")
print(matrix2)
print("\n–î–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—å:")
print(product)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 7
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º 3x2 —Ç–∞ 2x2, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—Ö –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏
–≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 8 —Ç–∞ –ø–µ—Ä–µ–º–Ω–æ–∂—Ç–µ —ó—Ö –º—ñ–∂ —Å–æ–±–æ—é.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –ø–µ—Ä—à—É –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x2 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 8
matrix1 = np.random.randint(1, 9, (3, 2))

# –°—Ç–≤–æ—Ä—é—î–º–æ –¥—Ä—É–≥—É –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 2x2 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 8
matrix2 = np.random.randint(1, 9, (2, 2))

# –ü–µ—Ä–µ–º–Ω–æ–∂—É—î–º–æ –º–∞—Ç—Ä–∏—Ü—ñ
product = np.dot(matrix1, matrix2)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–µ—Ä—à–∞ –º–∞—Ç—Ä–∏—Ü—è (3x2):")
print(matrix1)
print("\n–î—Ä—É–≥–∞ –º–∞—Ç—Ä–∏—Ü—è (2x2):")
print(matrix2)
print("\n–î–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—å (3x2 * 2x2):")
print(product)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 8
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—ó –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ
–≤—ñ–¥ 1 –¥–æ 8 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å —ó—ó –æ–±–µ—Ä–Ω–µ–Ω—É –º–∞—Ç—Ä–∏—Ü—é.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 8
matrix = np.random.randint(1, 9, (3, 3))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ –æ–±–µ—Ä–Ω–µ–Ω—É –º–∞—Ç—Ä–∏—Ü—é
inverse_matrix = np.linalg.inv(matrix)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–æ—á–∞—Ç–∫–æ–≤–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):")
print(matrix)
print("\n–û–±–µ—Ä–Ω–µ–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è:")
print(inverse_matrix)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 9
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 4x4, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—ó –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤
–¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ -1 –¥–æ 1 —Ç–∞ —Ç—Ä–∞–Ω—Å–ø–æ–Ω—É–π—Ç–µ —ó—ó.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 4x4 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -1 –¥–æ 1
matrix = np.random.uniform(-1, 1, (4, 4))

# –¢—Ä–∞–Ω—Å–ø–æ–Ω—É—î–º–æ –º–∞—Ç—Ä–∏—Ü—é
transposed_matrix = matrix.T

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–æ—á–∞—Ç–∫–æ–≤–∞ –º–∞—Ç—Ä–∏—Ü—è (4x4):")
print(matrix)
print("\n–¢—Ä–∞–Ω—Å–ø–æ–Ω–æ–≤–∞–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è:")
print(transposed_matrix)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 10
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x4 —Ç–∞ –≤–µ–∫—Ç–æ—Ä —Ä–æ–∑–º—ñ—Ä–æ–º 4x1, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é
–≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 0 –¥–æ 1, –∞ –≤–µ–∫—Ç–æ—Ä - –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤
–¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 10, —ñ –∑–Ω–∞–π–¥—ñ—Ç—å –¥–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x4 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 0 –¥–æ 1
matrix = np.random.uniform(0, 1, (3, 4))

# –°—Ç–≤–æ—Ä—é—î–º–æ –≤–µ–∫—Ç–æ—Ä —Ä–æ–∑–º—ñ—Ä–æ–º 4x1 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 10
vector = np.random.randint(1, 11, (4, 1))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ –¥–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä
result = np.dot(matrix, vector)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ú–∞—Ç—Ä–∏—Ü—è (3x4):")
print(matrix)
print("\n–í–µ–∫—Ç–æ—Ä (4x1):")
print(vector)
print("\n–î–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä:")
print(result)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 11
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 2x3 —Ç–∞ –≤–µ–∫—Ç–æ—Ä —Ä–æ–∑–º—ñ—Ä–æ–º 3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é
–≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 10, –∞ –≤–µ–∫—Ç–æ—Ä - –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏
—á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 0 –¥–æ 1, —ñ –∑–Ω–∞–π–¥—ñ—Ç—å –¥–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 2x3 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 10
matrix = np.random.randint(1, 11, (2, 3))

# –°—Ç–≤–æ—Ä—é—î–º–æ –≤–µ–∫—Ç–æ—Ä —Ä–æ–∑–º—ñ—Ä–æ–º 3 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 0 –¥–æ 1
vector = np.random.uniform(0, 1, 3)

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ –¥–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä
result = np.dot(matrix, vector)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ú–∞—Ç—Ä–∏—Ü—è (2x3):")
print(matrix)
print("\n–í–µ–∫—Ç–æ—Ä (3):")
print(vector)
print("\n–î–æ–±—É—Ç–æ–∫ –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ –≤–µ–∫—Ç–æ—Ä:")
print(result)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 12
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—Ö –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤
–¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 3 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å —Å—É–º—É –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥—Ä—É–≥–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 3
matrix1 = np.random.randint(1, 4, (3, 3))
matrix2 = np.random.randint(1, 4, (3, 3))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ —Å—É–º—É –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥—Ä—É–≥–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ
sum_matrix2 = np.sum(matrix2)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–µ—Ä—à–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):")
print(matrix1)
print("\n–î—Ä—É–≥–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):")
print(matrix2)
print("\n–°—É–º–∞ –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥—Ä—É–≥–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ:")
print(sum_matrix2)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 13
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—Ö –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤
–¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 1 –¥–æ 5 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å —Å—É–º—É —ó—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º 3x3 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 5
matrix1 = np.random.randint(1, 6, (3, 3))
matrix2 = np.random.randint(1, 6, (3, 3))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ —Å—É–º—É –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –ø–µ—Ä—à–æ—ó —Ç–∞ –¥—Ä—É–≥–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ
sum_matrix1 = np.sum(matrix1)
sum_matrix2 = np.sum(matrix2)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ü–µ—Ä—à–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):")
print(matrix1)
print("\n–°—É–º–∞ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –ø–µ—Ä—à–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ:")
print(sum_matrix1)

print("\n–î—Ä—É–≥–∞ –º–∞—Ç—Ä–∏—Ü—è (3x3):")
print(matrix2)
print("\n–°—É–º–∞ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥—Ä—É–≥–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ:")
print(sum_matrix2)

# –ó–∞–≥–∞–ª—å–Ω–∞ —Å—É–º–∞ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –æ–±–æ—Ö –º–∞—Ç—Ä–∏—Ü—å
total_sum = sum_matrix1 + sum_matrix2
print("\n–ó–∞–≥–∞–ª—å–Ω–∞ —Å—É–º–∞ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –æ–±–æ—Ö –º–∞—Ç—Ä–∏—Ü—å:")
print(total_sum)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 14
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—ó –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ
–≤—ñ–¥ 1 –¥–æ 10 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å —Å—É–º—É –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –Ω–∞ –≥–æ–ª–æ–≤–Ω—ñ–π –¥—ñ–∞–≥–æ–Ω–∞–ª—ñ.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ —Ü—ñ–ª–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ 1 –¥–æ 10
matrix = np.random.randint(1, 11, (3, 3))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ —Å—É–º—É –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –Ω–∞ –≥–æ–ª–æ–≤–Ω—ñ–π –¥—ñ–∞–≥–æ–Ω–∞–ª—ñ
sum_diagonal = np.trace(matrix)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ú–∞—Ç—Ä–∏—Ü—è (3x3):")
print(matrix)
print("\n–°—É–º–∞ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –Ω–∞ –≥–æ–ª–æ–≤–Ω—ñ–π –¥—ñ–∞–≥–æ–Ω–∞–ª—ñ:")
print(sum_diagonal)
```

```python
#–ó–∞–≤–¥–∞–Ω–Ω—è 15
'''
–°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3, –∑–∞–ø–æ–≤–Ω—ñ—Ç—å —ó—ó –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤
–¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ -1 –¥–æ 1 —Ç–∞ –∑–Ω–∞–π–¥—ñ—Ç—å –≤–µ–∫—Ç–æ—Ä-—Å—Ç–æ–≤–ø—á–∏–∫, —â–æ –º—ñ—Å—Ç–∏—Ç—å –¥–æ–±—É—Ç–æ–∫ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
–∫–æ–∂–Ω–æ–≥–æ —Å—Ç–æ–≤–ø—Ü—è –º–∞—Ç—Ä–∏—Ü—ñ.
'''
# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º 3x3 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏ –≤—ñ–¥ -1 –¥–æ 1
matrix = np.random.uniform(-1, 1, (3, 3))

# –ó–Ω–∞—Ö–æ–¥–∏–º–æ –¥–æ–±—É—Ç–æ–∫ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –∫–æ–∂–Ω–æ–≥–æ —Å—Ç–æ–≤–ø—Ü—è
column_products = np.prod(matrix, axis=0)

# –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ –≤ –≤–µ–∫—Ç–æ—Ä-—Å—Ç–æ–≤–ø—á–∏–∫
column_vector = column_products.reshape(-1, 1)

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
print("–ú–∞—Ç—Ä–∏—Ü—è (3x3):")
print(matrix)
print("\n–í–µ–∫—Ç–æ—Ä-—Å—Ç–æ–≤–ø—á–∏–∫ –∑ –¥–æ–±—É—Ç–∫–æ–º –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –∫–æ–∂–Ω–æ–≥–æ —Å—Ç–æ–≤–ø—Ü—è:")
print(column_vector)
```
